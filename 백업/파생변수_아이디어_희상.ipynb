    > 학습 지속율 : 활성화 일수(접속 일수 개념) / 강좌 내 활동기간
- 학생들이 총 활동기간동안, 자주 접속 했는가를 묻는 지표
- 운영팀: 학생들이 꾸준히 접속하고 있나? (학습 지속율)
- 사업팀: 꾸준히 접속한 사람들이 수료율이 높군 ! -> 어떤 인사이트로 해석할 지 추가 아이디어 필요

> 초기 참여도 : 활성화 일 수가 3일보다 적은데, 전체 이벤트 수가 0보다 큰 경우
- 학습 열정이 가장 높은 등록 후 초반에 어느단계까지 진입했는지 볼 수 있음.

> 학습 참여 점수 : 사용자 행동지표별로 가중치를 둬 점수화
- 학습 참여에 얼만큼 몰입했는지 점수로 수치화 가능
- 학습 참여율이 높으면 -> 수료율 높음 인사이트 가능

> 이탈 위험 학생 : 활성화 일수가 낮고, 영상재생 수 혹은 챕터를 절반 이상 탐색하지 않은 학생
- 이탈 위험이 있는 학생을 1로, 이탈 위험이 낮은 학생을 0으로 구분
- 운영팀: 이탈 위험이 높은 학생을 집중 관리 (이탈률을 낮추기 위한 전략으로 쓸 수 있음)
- 사업팀: 이탈 위험이 높은 학생이 많은 강의를 확인 (어떤 강의가 이탈자가 많이 나올지 예상 가능)

> 학습속도 : 활동 지속기간 중 탐색한 강좌 수
- registered 단계부터 explored 단계까지 전환되는 기준이 nchapters로 결정되기 때문에, 학습 속도를 나타내기에 좋은 지표가 될 것 같음

> 강의 시청비율 : 모든 이벤트 발생 수 중 영상시청의 비율
- 각 퍼널 단계로 비교해본다면 강의 품질 지표로 사용할 수 있지 않을까

> 수료 가능성 : 학습 지속율과 학습 참여 점수를 곱함
- 꾸준히 활동하면서 학습을 이어가고 있는지를 확인할 수 있는 지표
- 사업팀 : 어떤 강의가 학생들 수료 가능성이 높은지 확인할 수 있음 -> 나중에 유료인증제를 도입할 계획이 있다는 가정하에 이수율이 높은 강의를 판단한 후 예산을 분배할 수 있지 않을까

> 학습 참여도: 

```python
# 학습 지속율
pre["consistency"] = pre["ndays_act"] / pre["duration"]

# 초기 참여도(bool)
pre['early_engagement'] = pre[
    (pre['ndays_act'] <= 3)&
    (pre['nevents'] > 0)
    ].astype(int)

# 학습 참여 점수 (행동지표와 가중치는 상의 후 결정 필요. 근거있는 기준이 있으면 더 좋음)
pre['engagement_score'] = (
    pre['nevents'] * 0.4 +
    pre['ndays_act'] * 0.3 +
    pre['nplay_video'] * 0.2 +
    pre['nforum_posts'] * 0.1
)

# 이탈 위험 학생
pre['dropout_risk'] = (
    (pre['ndays_act'] < 2) |
    (pre['nplay_video'] == 0) |
    (pre['explored'] == 0)
).astype(int)

# 학습속도
pre['chapters_per_day'] = pre['nchapters'] / pre['duration']

# 강의 시청비율
pre['passive_ratio'] = pre['nplay_video'] / pre['nevents']

# 수료 가능성
pre['completion_proxy'] = (pre['engagement_score'] * pre['consistency'])

# 코스별 가장 많이 탐색한 학생의 챕터 수
pre['course_max_chapters'] = pre[pre['step'].isin(['explored', 'certified'])].groupby('course_id')['nchapters'].transform('max')

# 진도율
pre['chpater_progress_pct'] = pre['nchpater'] / pre['course_max_chapter']

# 
pre[''] = pre['engagement_score'] * pre['chapter_progress_pct']
```
